### 1. Python Numerical Data Types
<h4>Integer Values (int) - </h4>

As mentioned before python has three built in numerical data types, they are: Integers Values, Floating Point Values and Complex Values.

Integer values represent whole numbers with no decimal places. Such as 1, 2 and 69. While 1 is an ineger 1.0 is not. 
Bellow we will run the type() function on an integer value and it should return int. 

```python
type(1)
```

When dealing with large integers or large numerical values in general, it is often hard to read. 100000 is a lot harder to read than 100,000. However in python we cannot use commas in our integer values to help us out instead we can use under scores. This is another valid way of writing 100000, 100_000.
Also unlike other languages in python there is no limit to the size of an int which is great!

```python
type(100_000)
```

<h4>Float Values (float) - </h4>
Floating point values are percise numbers that can be accurate to 15 decimal places. There are three ways to represent floating point values in python, all three are shown bellow:

```python
1000000.0
```

```python
1_000_000.0
```

```python
1e6
```

The first two are similar/ exactly the same as describing int values, while the third makes use of something called **E notation** short for exponential notation, this notation is often used when values are too big and fill a screen. <br/>
If you wish to use E notation you take the number on the left and multiply it by 10 to the power of the number on the right.\ 
1e+5 = 100000
1e-5 = 0.00001

As you see above E notation can be used to represent both positive and negitive exponets.

Unlike integer values, floating points do have a max size, yet this depends on the system dealing with the program. If you go beyond this your program will return **inf** which stands for infinity, yet its type will still be float. And for values that exceed the lowest or minimum posible value the program will return **-inf**.
